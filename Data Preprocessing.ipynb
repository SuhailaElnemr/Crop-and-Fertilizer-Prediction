```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('Crop and fertilizer dataset.csv')
data.head()
```

```python
data.columns
```

```python
data.drop(labels = ['District_Name' , 'Link' ,'Rainfall'] , axis = 1 , inplace = True)
data.head()
```

```python
data['Soil_color'].unique()
```

```python
data['Soil_color'] = data['Soil_color'].str.strip(' ')
data['Soil_color'].unique()
```

```python
data['Crop'].unique()
```

```python
data['Fertilizer'].unique()
```

```python
data.duplicated().sum()
data.drop_duplicates()
```

```python
s = (data.dtypes =="object")
object_cols = list(s[s].index)
print("Categorical variables:")
print(object_cols)
print(len(object_cols))
```

```python
encoder = LabelEncoder()

for i in object_cols:
    data[i] = encoder.fit_transform(data[i])
    
data.dtypes =="object"
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(data.corr(), annot=True,fmt=".2f", cmap="viridis")
plt.show()
```

```python
data.to_csv('data.csv', index=False)
```
